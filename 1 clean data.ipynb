```python
import pandas as pd
import datetime as dt;
```

```python
data = '/kaggle/input/product-rating/product_rating.csv'
columns_to_read = ['rating', 'comment','date'] 
df=pd.read_csv(data, usecols=columns_to_read)
df.head()
```

```python
# extract rating
df['rating'] = df['rating'].str.extract(r'Rated (\d+(?:\.\d+)?)')

# extract date
df['date'] = df['date'].str.split(':', expand=True)[1].str.strip()
df.head()
```

```python
df['date'] = pd.to_datetime(df['date'], format='%B %d, %Y', errors='coerce')
df.head()
```

```python
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['date'] = df['date'].dt.day
df.head()
```

```python
df['rating'] = pd.to_numeric(df['rating'], errors='coerce').astype('Int64')
df.isna().sum()
```
